This notebook demonstrates a streamlined approach to fetch and visualize Sentinel‑2 spectral index tiles (e.g., NDVI, NDWI) using the **VirtuGhan** Python package. VirtuGhan’s `vcube.tile.TileProcessor` handles STAC search, COG streaming, custom index formulas, and styling with Matplotlib colormaps, all in one high‑level API. Results are stored in a `VirtualCube` container for easy listing, retrieval, and batch export.

````python
# 1) Imports
import nest_asyncio
nest_asyncio.apply()

from a_4_utils import fetch_index_tile, save_tile_image, VirtualCube
from IPython.display import display
import os
````

````python
# 2) Parameters (Bucharest example)
lat, lon      = 44.43, 26.10
zoom          = 12
start_date    = "2024-04-01"
end_date      = "2024-04-30"
cloud_cover   = 30
````

````python
# 3) Prepare output folder
output_folder = "A4_outputs_Results"
os.makedirs(output_folder, exist_ok=True)

# 4) Instantiate the VirtualCube
vc = VirtualCube()
````

````python
# 5) Fetch NDVI tile and add to cube
ndvi_img, ndvi_meta, ndvi_coords = fetch_index_tile(
    lat, lon, zoom,
    start_date, end_date,
    cloud_cover,
    band1="red", band2="nir",
    formula="(band2-band1)/(band2+band1)",
    colormap_str="RdYlGn"
)
vc.add_tile("NDVI", ndvi_img, ndvi_meta, ndvi_coords)

# 6) Fetch NDWI tile and add to cube
ndwi_img, ndwi_meta, ndwi_coords = fetch_index_tile(
    lat, lon, zoom,
    start_date, end_date,
    cloud_cover,
    band1="green", band2="nir",
    formula="(band1-band2)/(band1+band2)",
    colormap_str="BrBG"
)
vc.add_tile("NDWI", ndwi_img, ndwi_meta, ndwi_coords)
````

````python
# 7) List what’s in the cube
print("Stored tiles:", vc.list_tiles())

# 8) Display each tile inline
for (x,y,z,name) in vc.list_tiles():
    img, meta = vc.get_tile(name, (x,y,z))
    print(f"\n{name} tile at Z={z}, X={x}, Y={y}, Date={meta['properties']['datetime']}")
    display(img)

# 9) Save all stored tiles to disk
vc.save_all(output_folder)
````
